```python
import numpy as np
import torch                                    #torch library
from torch.utils.data import Dataset,DataLoader #helps in dividing the dataset into mini batches for trainimg
import glob                                     #helps in itertaing throuh the dataset
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix,accuracy_score
import cv2
```

```python
tumor = []
path='./brain_tumor_dataset/yes/*.jpg' # all yes-tumor .jpeg images
for f in glob.iglob(path):
    img = cv2.imread(f)
    img = cv2.resize(img,(120,120)) #make the images dimensionally equal(W,H)-(120,120)
    b,g,r = cv2.split(img)     #split the B,G,R part of the image(channels) lists
    img = cv2.merge([r,g,b]) #reordering the image 
    tumor.append(img)
```

```python
len(tumor)
```

```python
for img in tumor:
    print(img.shape)
```

```python
no_tumor = []
path='./brain_tumor_dataset/no/*.jpg' # all no-tumor .jpeg images
for f in glob.iglob(path):
    img = cv2.imread(f)
    img = cv2.resize(img,(120,120)) #make the images dimensionally equal(W,H)-(120,120)
    b,g,r = cv2.split(img)     #split the B,G,R part of the image(channels) lists
    img = cv2.merge([r,g,b]) #reordering the image 
    no_tumor.append(img)
```

```python
len(no_tumor)
```

```python
for img in no_tumor:
    print(img.shape)
```

```python
#generate random tumor and no_tumor images
def plot_random(no_tumor,tumor, num=6):
    no_tumor_images = no_tumor[np.random.choice(no_tumor.shape[0], num, replace=False)]
    tumor_images = tumor[np.random.choice(tumor.shape[0], num, replace=False)]
    
    plt.figure(figsize=(16,9))
    for i in range(num):
        plt.subplot(1,num,i+1)  #(row, column,in which row to show)
        plt.title("No_tumor")
        plt.imshow(no_tumor[i])
        
    plt.figure(figsize=(16,9))
    for i in range(num):
        plt.subplot(1,num,i+1)  #(row, column,in which row to show)
        plt.title("tumor")
        plt.imshow(tumor[i])
        
            
        
```
